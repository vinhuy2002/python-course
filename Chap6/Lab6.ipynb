```python
# 5.1
df = pd.DataFrame({
    'ord_no': [70001, np.nan, 70002, 70004, np.nan, 70005, '--'],
    'ord_date': ["2012-10-05", "2012-09-10", '?', "2012-08-17", "2012-09-10", "2012-07-27", "2012-09-10"],
    'customer_id': [3002,3001,3001,3003,3002, np.nan, 3001],
    'Payment': [120000, 15764, 30034, 64633, np.nan, 49593, 31399]
})
df
```

```python
print(df.isnull())
```

```python
print(df.notnull())
```

```python
print(df.isna().any())
```

```python
print("\nTotal missing values in a dataframe:")
tot_missing_vals = df.isnull().sum().sum()
print(tot_missing_vals)
```

```python
print("Number of missing values of the said dataframe:")
print(df.isna().sum())
```

```python
res = df.replace({"?": np.nan, "--": np.nan})
print(res)
```

```python
res = df['ord_no'].isnull().to_numpy().nonzero()
print(res)
```

```python
res = df.dropna(how='all')
print(res)
```

```python
res = df.dropna(inplace=False)
print(res)
```

```python
res = df.isna().sum().sum()
print("Total of missing values: ",res)
```

```python
res = df['ord_no'].fillna(0, inplace=False)
print(res)
```

```python
print("Using median in 'Payment' to replace NaN: ")
df['Payment'].fillna(df['Payment'].median(), inplace=True)
print(df)
print("Using mean to replace NaN:")
df['Payment'].fillna(int(df['Payment'].mean()), inplace=True)
print(df)
```
