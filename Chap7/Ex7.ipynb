```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10,10,200)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color="green", linestyle="", marker="x")
plt.show()
```

```python
plt.plot(x,y, color="green", marker="o", linestyle="")
plt.show()
```

```python
indexX = x[50:80]
indexY = np.sin(indexX)
plt.plot(indexX, indexY)
plt.show()
```

```python
indexX = x[51:56]
indexY = np.sin(indexX)
plt.plot(indexX, indexY)
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv("weight-height.csv", engine='python')
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["Height"].values, df["Weight"].values)
plt.show()
```

```python
maleArr = df.loc[df["Gender"] == "Male"]
femaleArr = df.loc[df["Gender"] == "Female"]
plt.scatter(maleArr["Height"].values, maleArr["Weight"].values, color="blue")
plt.scatter(femaleArr["Height"].values, femaleArr["Weight"].values, color="red")
plt.show()
```

```python
data1 = pd.read_csv("company-sales.csv", engine="python")
df1= pd.DataFrame(data1)
```

```python
df1
```

```python
plt.plot(df1['month_number'], df1['total_profit'], color="blue")
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.show()
```

```python
plt.plot(df1['month_number'], df1['total_profit'], color="red", label='Profit data of last year', linestyle="dotted", marker=".",linewidth=3, markersize= 12)
plt.title("Company profit per month")
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.legend(loc=4)
plt.show()
```
