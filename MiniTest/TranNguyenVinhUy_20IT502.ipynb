```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("sale_data_vn.txt")
df = pd.DataFrame(data)
df
```

```python
plt.plot(df["City"], df["Gender"])
plt.show()
```

```python
HaNoi = df.loc[df["City"] == "Hà Nội"]
TPHCM = df.loc[df["City"] == "TP HCM"]
DaNang = df.loc[df["City"] == "Đà Nẵng"]
```

```python
HaNoiNam = HaNoi.loc[HaNoi["Gender"] == "Nam"]
HaNoiNu = HaNoi.loc[HaNoi["Gender"] == "Nữ"]
TPHCMNam = TPHCM.loc[TPHCM["Gender"] == "Nam"]
TPHCMNu = TPHCM.loc[TPHCM["Gender"] == "Nữ"]
DaNangNam = DaNang.loc[DaNang["Gender"] == "Nam"]
DaNangNu = DaNang.loc[DaNang["Gender"] == "Nữ"]
```

```python
labels = ["Hà Nội - Nam", "Hà Nội - Nữ", "TPHCM - Nam", "TPHCM - Nữ", "Đà Nẵng - Nam", "Đà Nẵng - Nữ"]
sizes = [len(HaNoiNam), len(HaNoiNu), len(TPHCMNam), len(TPHCMNu), len(DaNangNam), len(DaNangNu)]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```
